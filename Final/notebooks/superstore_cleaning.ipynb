```
import pandas as pd
```

```
df = pd.read_csv("Superstore_clean.csv")
```

```
df.columns = (
    df.columns
    .str.strip()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)
```

```
dim_customer = df[['Customer_ID', 'Customer_Name', 'Segment', 'Region']].drop_duplicates()

dim_product = df[['Product_ID', 'Product_Name', 'Category',
                   'Sub_Category']].drop_duplicates()

fact_sales = df[['Order_ID', 'Order_Date', 'Ship_Date', 'Ship_Mode',
                  'Customer_ID', 'Product_ID', 'Sales', 'Quantity',
                  'Discount', 'Profit', 'City', 'State', 'Postal_Code']]
```

```
dim_customer.to_csv("dim_customer.csv", index=False)
dim_product.to_csv("dim_product.csv", index=False)
fact_sales.to_csv("fact_sales.csv", index=False)
```
